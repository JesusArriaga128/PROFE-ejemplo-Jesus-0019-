```python
class Automovil:
    def __init__(self, marca, placas, precio):
        self.marca = marca
        self.placas = placas
        self.precio = precio

    def leerdatos(self):
        self.marca = input("Marca: ")
        self.placas = input("Placas: ")
        self.precio = input("Precio: ")

    def obtener_formato(self):

        resumen = f"Auto: {self.marca} | Placas: {self.placas} | Precio: ${self.precio}"
        return resumen

    def mostrardatos(self):

        print(self.obtener_formato())

print("Carro numero 1")
carro1 = Automovil("", "", "")
carro1.leerdatos()
carro1.mostrardatos()

print("*_*_*_*_*_*_*_*_*_*_*__*_*_* ")

print("Carro numero 2")
carro2 = Automovil("", "", "")
carro2.leerdatos()
carro2.mostrardatos()

print("*_*_*_*_*_*_*_*_*_*_*__*_*_* ")

print("Carro numero 3")
carro3 = Automovil("", "", "")
carro3.leerdatos()
carro3.mostrardatos()

print("Programa realizado por Jesus Arriaga NL = 5")
```
